# 사전학습 언어 모델 파인튜닝 및 배포

> 소개: 이 섹션에서는 사전학습 모델의 파인튜닝을 다룹니다.
특정 작업에서 사전학습 모델의 성능을 향상시키고 싶나요? 적절한 사전학습 모델을 선택하고, 특정 작업에 파인튜닝한 후, 파인튜닝된 모델을 편리하게 사용할 수 있는 Demo로 배포해 봅시다!
## 이 튜토리얼의 목표:
1. Transformers 라이브러리 사용법 익히기
2. 사전학습 모델의 파인튜닝 및 추론 방법 습득 (모듈 분리 커스텀 버전 & 기본 통합 버전)
3. Gradio Spaces를 활용한 Demo 배포 방법 습득
4. 다양한 유형의 사전학습 모델 선택 기준 및 적용 시나리오 이해

### 1. 사전 준비:
#### 1.1 라이브러리 소개: Transformers
https://github.com/huggingface/transformers

> 🤗 Transformers는 최신 사전학습 모델을 쉽게 다운로드하고 훈련할 수 있는 API와 도구를 제공합니다. 사전학습 모델을 사용하면 계산 비용과 탄소 배출을 줄이고, 처음부터 훈련하는 데 필요한 시간과 자원을 절약할 수 있습니다. 이 모델들은 다양한 모달리티에서 일반적인 작업을 지원합니다:
📝 자연어처리: 텍스트 분류, 개체명 인식, 질의응답, 언어 모델링, 요약, 번역, 다지선다, 텍스트 생성.
🖼️ 컴퓨터 비전: 이미지 분류, 객체 탐지, 의미 분할.
🗣️ 오디오: 자동 음성 인식, 오디오 분류.
🐙 멀티모달: 테이블 질의응답, 광학 문자 인식(OCR), 스캔 문서에서 정보 추출, 비디오 분류, 시각적 질의응답.

#### 1.2 환경 설치: 텍스트 분류(예: 가짜 뉴스 탐지)를 예시로

https://github.com/huggingface/transformers/tree/main/examples/pytorch/text-classification

2. 환경 설치:
- 1. conda로 새 환경 생성: `conda create -n llm python=3.9`
- 2. 가상 환경 활성화: `conda activate llm`
- 3. `pip install transformers`
- 4. requirements.txt에서 자동 설치되는 torch를 제거한 후, `pip install -r requirements.txt` 실행

> 다운로드 속도가 느린 경우, 국내 미러를 사용할 수 있습니다: `pip [패키지명] -i https://pypi.tuna.tsinghua.edu.cn/simple`

> 국내 미러로 pytorch를 설치하면 CPU 버전이 자동으로 선택되어 GPU 실행이 불가능할 수 있으므로——

- 5. `conda install pytorch`

> 다운로드 속도가 느린 경우, 해당 블로그를 참고하여 conda 미러를 설정하세요: https://blog.csdn.net/weixin_42797483/article/details/132048218

#### 3.2 모델 훈련
검증 세트로 동시에 검증하고, 테스트 세트로 예측을 수행하려면 아래 스크립트를 실행하세요:

```shell
python run_classification.py \
    --model_name_or_path  bert-base-uncased \
    --train_file data/train.csv \
    --validation_file data/val.csv \
    --test_file data/test.csv \
    --shuffle_train_dataset \
    --metric_name accuracy \
    --text_column_name "text" \
    --text_column_delimiter "\n" \
    --label_column_name "target" \
    --do_train \
    --do_eval \
    --do_predict \
    --max_seq_length 512 \
    --per_device_train_batch_size 32 \
    --learning_rate 2e-5 \
    --num_train_epochs 1 \
    --output_dir experiments/
```

오류가 발생하거나 멈추는 경우, 일반적으로 네트워크 문제입니다:

1. <u>모델 다운로드 시 "Network is unreachable" 오류가 발생하면, 모델을 로컬에 수동으로 다운로드하세요: https://huggingface.co/google-bert/bert-base-uncased</u>
2. <u>데이터를 로딩한 후 멈추고, CTRL+C로 종료하면 "connection"에서 멈춘 것으로 표시되는 경우, evaluate 패키지가 평가 지표를 로딩할 때 네트워크 연결에 실패한 것입니다.</u>

![bug](./assets/9.png)

이 경우, evaluate의 GitHub에서 전체 패키지를 다운로드하세요: https://github.com/huggingface/evaluate/tree/main, 그리고 하이퍼파라미터의 --metric_name 경로를 로컬 지표 경로로 변경하세요:

```shell
python run_classification.py \
    --model_name_or_path  bert-base-uncased \
    --train_file data/train.csv \
    --validation_file data/val.csv \
    --test_file data/test.csv \
    --shuffle_train_dataset \
    --metric_name evaluate/metrics/accuracy/accuracy.py \
    --text_column_name "text" \
    --text_column_delimiter "\n" \
    --label_column_name "target" \
    --do_train \
    --do_eval \
    --do_predict \
    --max_seq_length 512 \
    --per_device_train_batch_size 32 \
    --learning_rate 2e-5 \
    --num_train_epochs 1 \
    --output_dir experiments/
```
